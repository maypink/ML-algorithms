```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data/Summary_of_Weather.csv')
```

```python
X = data.MinTemp.values.reshape(-1, 1)
y = data.MaxTemp
```

```python
plt.scatter(X, y)
```

```python
from linreg import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(X, y)
```

```python
pred = reg.predict(X)
```

```python
plt.scatter(X, y)
plt.plot(X, pred, c='red')
```

```python
from sklearn.metrics import mean_squared_error
```

```python
print('MSE: ', mean_squared_error(y, pred))
```

Sklearn Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
sklearn_reg = LinearRegression().fit(X, y)
```

```python
sklearn_pred = sklearn_reg.predict(X)
```

```python
plt.scatter(X, y)
plt.plot(X, sklearn_pred, c='red')
```

```python
print('MSE: ', mean_squared_error(y, sklearn_pred))
```
