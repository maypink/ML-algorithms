```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data/Summary_of_Weather.csv')
```

```python
X = data.MinTemp.values.reshape(-1, 1)
y = data.MaxTemp
```

```python
plt.scatter(X, y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
import pylab
```

```python
pylab.subplot (2, 2, 1)
plt.scatter(X_train, y_train)
pylab.subplot (2, 2, 2)
plt.scatter(X_test, y_test)
```

```python
from linreg import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(X_train, y_train)
```

```python
pred = reg.predict(X_test)
```

```python
plt.scatter(X, y)
plt.plot(X_test, pred, c='red')
```

```python
from sklearn.metrics import mean_squared_error
```

```python
print('MSE: ', mean_squared_error(y_test, pred))
```

Sklearn Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
sklearn_reg = LinearRegression().fit(X_train, y_train)
```

```python
sklearn_pred = sklearn_reg.predict(X_test)
```

```python
plt.scatter(X, y)
plt.plot(X_test, sklearn_pred, c='red')
```

```python
print('MSE: ', mean_squared_error(y_test, sklearn_pred))
```
